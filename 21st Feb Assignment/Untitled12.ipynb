```python
page= requests.get('https://letterboxd.com/matthew/list/box-office-mojo-all-time-worldwide/detail/')
page
```

```python
Soup= BeautifulSoup(page.content)
Soup
```

```python
Title= Soup.find('h2', class_="headline-2 prettify")
Title
```

```python
Title.text
```

```python
Title.text.replace('2009','')
```

```python
Title=[]
for i in Soup.find_all('h2', class_="headline-2 prettify"):
    Title.append(i.text)
Title
```

```python
Rating= Soup.find('span', class_="rating rated-7")
Rating
```

```python
Rating.text
```

```python
Rating= []
for i in Soup.find_all('span', class_="rating rated-7"):
    Rating.append(i.text)
Rating
```

```python
Release= Soup.find('small',class_="metadata")
Release
```

```python
Release.text
```

```python
Release=[]
for i in Soup.find_all('small',class_="metadata"):
    Release.append(i.text)
Release
```

```python
import pandas as pd
```

```python
df=pd.DataFrame({'Title':Title, 'Rating':Rating, 'Release':Release})
print(df)
```
